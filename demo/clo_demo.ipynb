## Deep Curriculum Learning Optimization (deepclo)



```python
import matplotlib.pyplot as plt

from deepclo.algorithms.curriculum import Curriculum
from deepclo.core.measures.measure_functions import Measure
from deepclo.pipe.dataset import ImageDataProvider
from deepclo.utils import show_images
```

```python
dataset = ImageDataProvider(dataset_name='cifar10')

label = dataset.y_train[0]
batch_size = 8

train_batch = dataset.x_train[:batch_size]
train_label = dataset.y_train[:batch_size]

print(dataset)

curriculum = Curriculum(train_batch, train_label)

```

```python
curriculum.print_syllabus()
```

```python
batch_syllabus, ranks = curriculum.syllabus(measure=Measure.MI, reference_imag_index=None)
curriculum.print_syllabus()
```

```python
batch_syllabus, ranks = curriculum.syllabus(measure=Measure.ENTROPY, reference_imag_index=None)
curriculum.print_syllabus()
```
